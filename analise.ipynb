```python
import numpy as np
```

Carregando dados com tratamento de encoding e colunas específicas.

```python
#carregamdo o database, pegando apenas as tabelas Quantidade e preco
#vendas sem preço foram zeradas
data = np.genfromtxt("sales_data_sample.csv", delimiter=",",skip_header=1, usecols=(1, 2), filling_values=0, encoding="latin1")
```

Limpando dados para garantir que não existam quantidades ou preços negativos.

```python
#limpando dados nulos e negativos
filtro_quantidade = data[:, 0] > 0
filtro_preco = data[:, 1] > 0

filtro_vendas = filtro_quantidade & filtro_preco
```

```python
data_limpo = data[filtro_vendas]
```

```python
#multiplicando a quantidade pelo preco
faturamento = data_limpo[:, 0] * data_limpo[:, 1]
```

Cálculo de métricas financeiras principais (KPIs).

```python
#estatisticas
print(f"Faturamento total: {faturamento.sum():,.2f}")
print(f"Média de valor por venda: {faturamento.mean():,.2f}")
print(f"Maior venda registrada: {faturamento.max():,.2f}")
```
